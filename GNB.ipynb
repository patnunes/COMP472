```python
train_set2 = pd.read_csv("Assig1-Dataset/train_2.csv", header=None)
```

```python
train_set2.head()
```

```python
target_train = train_set2[1024] #y_train, the set of labels to all the data in x_train
```

```python
x_train = train_set2.drop(1024,axis=1) #the training data set
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
dtree = DecisionTreeClassifier(criterion = "entropy")
```

```python
dtree = dtree.fit(x_train, target_train)
```

```python
y_pred = dtree.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
